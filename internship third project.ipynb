# Visualising the data using countplots

# Comment:
#### 1.Maximum dealings happened in Region- 'West' and Segment- 'Consumer'.
#### 2.Maximum dealings happened in Region- 'West' and Quantity- '3'
#### 3.Minimum dealings happened in Region- 'South' and Segment- 'Home Office'

# State-wise Analysis

```python
df=data['State'].value_counts()
df
```

```python
df.plot.bar(title="state wise dealings",figsize=(15,7));
plt.xlabel("States");
plt.ylabel("No. of deals");
plt.show();
```

# Comment:
#### 1. State- 'California' has the highest number of deals.
#### 2. State-'North Dakota' has the lowest number of deals.

```python
df1=df.mean()
df1
```

```python
df2=df[df<df1]
df2
```

#### Average no. of deals per state is 204(approximately). Therefore, the States having no. of deals below 204 should have to improve their business strategy and gain the interests of their customers by providing more discounts. For profit, they have to increase their no. of deals.

```python
df_state=data.groupby(['State'])[['Sales','Profit','Gross Profit','Discount']].mean()
df_state
```

```python
df_state['Sales'].plot.bar(title="State-wise Sales Analysis", figsize=(12,6) , color='r');
plt.show();
```

# Comment:
#### 1. The highest number of sales is in 'Wyoming' State.
#### 2. The lowest number of sales is in 'South Dakota' State.

# Profit Analysis

```python
df_state['Profit'].plot.bar(title="State-wise Profit Analysis", figsize=(12,6) , color='g');
plt.show();
```

# Comment:
#### 1. The highest amount of profit is in 'Vermont' State.
#### 2. The lowest amount of profit are in 'Ohio' and 'Colorado' State.

```python
plt.figure(figsize=(15,8));
data.groupby(['State']).sum()['Gross Profit'].plot.bar(title=" Plotting the Gross Profits, State wise");

plt.xticks(rotation=90);
plt.ylabel("Gross Profit");
```

# Comment:
#### 1. The highest amount of Gross-Profit is in 'California' State.
#### 2. The lowest amount of Gross-Profit is in 'Texas' State.

```python
plt.figure(figsize=(10,5));
data.groupby(['City']).sum()['Gross Profit'].sort_values(ascending=False).head(30).plot.bar(title=" Plotting 30 Cities with highest Gross Profits");
plt.ylabel("Gross Profit");
```

# Comment:
#### The highest City with the Gross-Profit is 'New York City'.
